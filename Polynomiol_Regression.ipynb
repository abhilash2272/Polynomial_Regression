```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/Position_Salaries.csv')
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)

# Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
plt.plot(x,lin_reg.predict(x),color='blue')
plt.scatter(x,y,color='red')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.title('Truth or Bluff (Linear Regression)')
```

```python
# Visualising the Polynomial Regression results
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(poly_reg.fit_transform(x)), color = 'green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
linear_prediction=lin_reg.predict([[6.5]])
print("Linear regression for 6.5 level:",linear_prediction)
```

```python
poly_prediction=lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print("Polynomial regression for level 6.5",poly_prediction)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
y_pred=lin_reg.predict(x_test)
print("R2 score for linear regression:",r2_score(y_test,y_pred))
print("Mean square error",mean_squared_error(y_test,y_pred))
print("Mean absolute error",mean_absolute_error(y_test,y_pred))
print("Root mean square error",root_mean_squared_error(y_test,y_pred))
```

```python
y_poly_pred=lin_reg_2.predict(poly_reg.fit_transform(x_test))
print("R2 score for polynomial regression:",r2_score(y_test,y_poly_pred))
print("Mean square error",mean_squared_error(y_test,y_poly_pred))
print("Mean absolute error",mean_absolute_error(y_test,y_poly_pred))
print("Root mean square error",root_mean_squared_error(y_test,y_poly_pred))
```
